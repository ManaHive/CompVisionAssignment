```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
```

```python
train_df = pd.read_csv("training_features.csv")
test_df = pd.read_csv("test_features.csv")

train_df
```

```python
train_metadata = pd.read_csv("training data/metadata.csv")
test_metadata = pd.read_csv("test data/metadata.csv")

train_metadata.info()
```

```python
train_df.describe()
```

```python
test_df.describe()
```

```python
train_df.drop(columns=['lbp_2','lbp_4','lbp_6'], inplace=True)
test_df.drop(columns=['lbp_2','lbp_4','lbp_6'], inplace=True)

train_df.rename(columns={
    'lbp_3':'lbp_2',
    'lbp_5':'lbp_3',
    'lbp_7':'lbp_4',
    'lbp_8':'lbp_5'
}, inplace=True)

test_df.rename(columns={
    'lbp_3':'lbp_2',
    'lbp_5':'lbp_3',
    'lbp_7':'lbp_4',
    'lbp_8':'lbp_5'
}, inplace=True)
train_df.describe()
```

```python
# checking for null values
train_df.info()
```

```python
train_df_merged = train_df.join(train_metadata.loc[:,"benign_malignant"], how='left')
test_df_merged = test_df.join(test_metadata.loc[:,"benign_malignant"], how='left')
train_df_merged.info()
```

```python
test_df_merged.info()
```

```python
X_train = train_df_merged.drop(columns="benign_malignant")
X_test = test_df_merged.drop(columns="benign_malignant")

y_train = train_df_merged.loc[:,"benign_malignant"]
y_test = test_df_merged.loc[:,"benign_malignant"]

X_train
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"KNN Accuracy (Before Tuning): {accuracy * 100:.2f}%")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid_knn ={
    "n_neighbors": [1,3,5,8,10,30],
    "leaf_size": [1,5,10,20,30],
    "p":[1,2]
}

grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, n_jobs=1, scoring="accuracy")
grid_search_knn.fit(X_train,y_train)

best_params_knn = grid_search_knn.best_params_
print("Best Hyperparameters (KNN):", best_params_knn)

y_pred_knn = grid_search_knn.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"K-Nearest Neightbors Accuracy (After Tuning): {accuracy_knn*100:.2f}%")
print(classification_report(y_test,y_pred))
```
